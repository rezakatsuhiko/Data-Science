```python
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imshow('data/roger-federer.jpg')
```

```python
img = imread('data/roger-federer.jpg')
```

```python
img.shape
```

```python
img = img/255
```

```python
imshow(img)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(img.reshape(-1,3))
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
img_new = kmeans.cluster_centers_[kmeans.labels_]
```

```python
img_tf = img_new.reshape(683,1024,3)
```

```python
imshow(img_tf)
```
